```python
#Librerias
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the cv2_imshow function
```

```python
#Imagen 1: Definir lados
W = 200
H = 200
#Matriz de medidas W, H
im = np.zeros((W, H), np.uint8)

# Dibujar lnea en la diagonal de la imagen con un ciclo for
for i in range(min(W, H)):
    im[i, i] = 255

# Mostrar la imagen
cv2_imshow(im)
```

```python
#Imagen 2
# Dimensiones de la imagen
W = 200
H = 200

# Crear una imagen en blanco
im2 = np.zeros((W, H, 3), np.uint8)

# Definir los colores
colors = [(0, 0, 0), (255, 0, 0), (0, 255, 0), (0, 0, 255)]

# Dibujar los cuadrados
for i, color in enumerate(colors):
    # Calcular el tamao del cuadrado
    size = (i + 1) * 20

    # Calcular las coordenadas del cuadrado
    start = (W // 2 - size // 2, H // 2 - size // 2)
    end = (W // 2 + size // 2, H // 2 + size // 2)
    # Dibujar el cuadrado
    cv2.rectangle(im2, start, end, color, 10)

# Mostrar la imagen
cv2_imshow(im2)
```

```python
#Imagen 3
W = 200
H = 200

im3 = np.zeros((H, W), np.uint8)

# Definir el nmero de lneas horizontales (y su frecuencia) basadas en la funcin seno
num_lines = 10
frequency = 20  # Frecuencia de la onda seno

# Dibujar las lneas horizontales que siguen una funcin seno
for i in range(num_lines):
    amplitude = int(H / (50))  # La amplitud de la lnea vara con la frecuencia
    y_offset = H // (num_lines + 1) * (i + 1)
    # Generar puntos basados en la funcin seno
    x = np.arange(0, W)
    y = amplitude * np.sin(2 * np.pi * frequency * x / W) + y_offset

    # Convertir los puntos a enteros
    points = np.column_stack((x, y)).astype(np.int32)

    # Dibujar la lnea
    cv2.polylines(im3, [points], isClosed=False, color=(255,0,0) , thickness=1)

# Mostrar la imagen
cv2_imshow(im3)
```
